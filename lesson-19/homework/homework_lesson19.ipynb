Homework Assignment 1: Analyzing Sales Data

```python
import pandas as pd

data=pd.read_csv('sales_data.csv')

df=pd.DataFrame(data)
df.head(2)
```

1.Group the data by the Category column and calculate the following aggregate statistics for each category:
Total quantity sold.
Average price per unit.
Maximum quantity sold in a single transaction.

```python
df.groupby('Category')['Quantity'].sum()
```

```python
df.groupby('Category').apply(lambda x:(x['Price']*x['Quantity']).sum()/x['Quantity'].sum()) 
```

2.Identify the top-selling product in each category based on the total quantity sold.

The first method.

```python
top_sellers = df.loc[df.groupby('Category')['Quantity'].idxmax()]
print(top_sellers)
```

The second method

```python
a=df['Category'].unique()

for x in a:
    print(df[df['Category']==x].nlargest(1,'Quantity'))
```

The second method

```python
df['Total']=(df['Quantity']*df['Price'])
df[df['Total']==df['Total'].max()]['Date']
```

Homework Assignment 2: Examining Customer Orders

```python
data=pd.read_csv('customer_orders.csv')
customers_orders=pd.DataFrame(data)
customers_orders
```

1.Group the data by CustomerID and filter out customers who have made less than 20 orders.

```python
filtered_customers_orders=customers_orders.groupby('CustomerID').filter(lambda x:len(x)>=20)
filtered_customers_orders
```

2.Identify the top-selling product in each category based on the total quantity sold.

```python
avarage_price=customers_orders.groupby('CustomerID')['Price'].mean()
more_120=avarage_price[avarage_price>120].index
customers_orders[customers_orders['CustomerID'].isin(more_120)]
```

3.Find the date on which the highest total sales (quantity * price) occurred.

```python
totals=customers_orders.groupby('Product').agg({'Quantity':'sum','Price':'sum'})
more_quantity_products=totals[totals['Quantity']>=5]
customers_orders[customers_orders['Product'].isin(more_quantity_products.index)]
```

Homework Assignment 3: Population Salary Analysis

1."task\population.db" sqlite database has population table.

```python
import sqlite3 

import pandas as pd

conn=sqlite3.connect('population.db')

df=pd.read_sql_query('SELECT  * FROM population',conn)
df
```

2."task\population salary analysis.xlsx" file defines Salary Band categories.
Read the data from population table and calculate following measures:

```python
data=pd.read_excel('population_salary_analysis (1).xlsx')
```

```python
bins = [0, 200000, 400000, 600000, 800000, 1000000, 1200000, 1400000, 1600000, 1800000, float('inf')]
labels = [
    'till $200,000',
    '$200,001 - $400,000',
    '$400,001 - $600,000',
    '$600,001 - $800,000',
    '$800,001 - $1,000,000',
    '$1,000,001 - $1,200,000',
    '$1,200,001 - $1,400,000',
    '$1,400,001 - $1,600,000',
    '$1,600,001 - $1,800,000',
    '$1,800,001 and over'
]
```

```python
df['Salary Band'] = pd.cut(df['salary'], bins=bins, labels=labels)
summary=df.groupby('Salary Band').agg(
    Number_of_population=('salary','count'),
    Avarage_Salary=('salary','mean'),
    mediana_salary=('salary','median'))
summary['Percentage']=summary['Number_of_population']/len(df)*100
summary.to_excel('My_resulr.xlsx')
```

3.Calculate the same measures in each State

```python

summary=df.groupby(['Salary Band','state']).agg(
    Number_of_population=('salary','count'),
    Avarage_Salary=('salary','mean'),
    mediana_salary=('salary','median'))
summary['Percentage']=summary['Number_of_population']/len(df)*100
summary.to_excel('My_result_by_state.xlsx')
```
